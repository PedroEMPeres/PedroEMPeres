```python
#FUNCAO PARA PEGAR O VENDEDOR, COM TRATAMENTO DE ERROS.
def pega_vendedor():
  #List compreension + zip para trazer todos os vendedores e imprimir, facilitando o caixa a achar o vendedor
  for valor in [f'{id} | {nome}' for id,nome in zip(list(DIM_VENDEDOR.keys()),list(DIM_VENDEDOR.values())) ]:
    print(valor)

  #Tratamento de erros.
  try:
    vendedor = int(input('Qual o Código do Vendedor acima realizou a venda? \n\n'))
  except:
    print('Digite  código do vendedor, não o nome ou qualquer outra coisa.\n\n')
    vendedor = pega_vendedor()

  vendedor = valida_vendedor(vendedor)
  return vendedor

#Checando se o vendedor inputado existe na lista de vendedor, se não, pergunto se quer cadastrar um novo
def valida_vendedor(vendedor):
  valida = True
  while valida == True:
    if vendedor not in DIM_VENDEDOR:
      print('Vendedor Não cadastrado. Os vendedores são:\n\n')
      for valor in [f'{id} | {nome}' for id,nome in zip(list(DIM_VENDEDOR.keys()),list(DIM_VENDEDOR.values())) ]:
        print(valor)

      cad_vendedor = input('Deseja Cadastrar Novo Vendedor? (S/N)\n\n')

      if cad_vendedor.upper() == 'S':
        vendedor = cadastra_vendedor()

      elif cad_vendedor.upper() == 'N':
        vendedor = valida_vendedor( pega_vendedor()) #recursivo
    else:
      valida = False
  return vendedor
#FUNCAO PARA PEGAR O CLIENTE, COM TRATAMENTO DE ERROS.
def pega_cliente():
  try:
    cliente = int(input('Qual o CPF do Cliente? \n\n'))
  except:
    print('Digite o CPF do Cliente de 11 algarismos (sem . ou - ), não o nome ou qualquer outra coisa.\n\n')
    cliente = pega_cliente()

  #breve validação do CPF, nao apliquei um validador real, apenas para qtd de algarismos >11
  cliente = str('{:0>11}'.format(cliente))
  cliente = valida_cliente(cliente)
  return cliente

#Validando se o cliente existe ou se será necessário cadastrar novo cliente
def valida_cliente(cliente):
  valida = True
  while valida == True:
    if cliente not in DIM_CLIENTE:
      print('Cliente Não cadastrado.\n\n')
      cad_cliente = input('Deseja Cadastrar Novo Cliente? (S/N)\n\n')

      if cad_cliente.upper() == 'S':
        cliente = cadastra_cliente()
        return cliente
      elif cad_cliente.upper() == 'N':
        cliente = '999'
        return cliente
    else:
      valida = False
  return cliente

#FUNCAO PARA PEGAR O PRODUTO, COM TRATAMENTO DE ERROS.
def pega_produto():
  try:
    produto = int(input('Qual o Código do produto? \n\n'))
  except:
    print('Digite  código do produto, não o nome ou qualquer outra coisa.\n\n')
    produto = pega_produto()

  produto = valida_produto(produto)
  return produto

#Checando se o produto existe ou se será necessário cadastrar novo produto
def valida_produto(produto):
  valida = True
  while valida == True:
    if produto not in DIM_PRODUTO:
      print('Produto Não cadastrado. \n\n')

      cad_produto = input('Deseja Cadastrar Novo Produto? (S/N)\n\n')

      if cad_produto.upper() == 'S':
        produto = cadastra_produto()

      elif cad_produto.upper() == 'N':
        produto = valida_produto( pega_produto()) #recursivo

    else:
      valida = False
  return produto

#FUNCAO PARA PEGAR O ANO, COM TRATAMENTO DE ERROS.
def pega_ano():
  print('Digite um ano válido a partir de 2022')
  try:
    ano = int(input('Digite o Ano: '))
  except:
    print('Digite um ano válido a partir de 2022')
    pega_ano()
  if ano > 2023 or ano <2022:
    ano = pega_ano()
  print(ano)
  return ano

#FUNCAO PARA PEGAR O MES, COM TRATAMENTO DE ERROS.
def pega_mes():
  print('Digite o mês em formato numérico ( Ex.: 01 para Janeiro, 02 para Fevereiro e etc.)')
  mes = input('Digite o mês: ')
  lista_mes=['01','02','03','04','05','06','07','08','09','10','11','12']
  if mes not in lista_mes:
    print('Formato inválido ou mês inexistente')
    mes = pega_mes()

  return mes

#FUNCAO PARA PEGAR O DIA, COM TRATAMENTO DE ERROS.
def pega_dia():

  try:
    dia = int(input('Digite o dia da compra: (dia do Mês de 01 a 31)'))
  except:
    print('Você digitou algo inválido.')
    dia = pega_dia()
  if dia <1 or dia > 31:
    print('Você digitou um valo de dia incoerente')
    dia = pega_dia()
  return str(dia)

#FUNCAO PARA PEGAR A QTD, COM TRATAMENTO DE ERROS.
def pega_qtd():
  try:
    qtd = int(input('Qual a quantidade do item'))
  except:
    print('Quantidade inválida')
    qtd = pega_qtd()


  return qtd

#FUNCAO PARA cadastrar as vendas e incluir na TABELA_FATO
def cadastra_venda(vendedor, cliente, produto, qtd, data, obs):
  #pego o ultimo id venda e acrescento 1 para o novo id
  id_venda = [ids['ID_VENDA'] for ids in TABELA_FATO][-1] + 1

  TABELA_FATO.append({'ID_VENDA':id_venda,'ID_PRODUTO':produto,'QTD_VENDIDA':qtd,'ID_VENDEDOR':vendedor,'ID_CLIENTE':cliente,'DTA_VENDA':data,'OBS':obs})

  #O loop abaixo serve para incluir mais de um produto no mesmo ID Venda
  mais_produtos = True
  while mais_produtos == True:
    mesma_venda = input('Deseja cadastrar outro item na mesma venda? (S/N)')
    if mesma_venda.upper() not in ['S','N']:
      print('Digite uma opção válida')
      continue
    elif mesma_venda.upper() == 'S':
      produto = pega_produto()
      qtd = pega_qtd()
      TABELA_FATO.append({'ID_VENDA':id_venda,'ID_PRODUTO':produto,'QTD_VENDIDA':qtd,'ID_VENDEDOR':vendedor,'ID_CLIENTE':cliente,'DTA_VENDA':data,'OBS':obs})
      continue
    else:
      mais_produtos = False

  return

#FUNCAO PARA PEGAR O ID_VENDA, COM TRATAMENTO DE ERROS.
def pega_id_venda():
  try:
    id_venda = int(input('Qual o id da venda?'))
  except:
    print('Formato de ID Inválido')
    pega_id_venda()
  if id_venda not in [ids['ID_VENDA'] for ids in TABELA_FATO]:
    print('ID não encontrado')
    pega_id_venda()

  return id_venda

#FUNCAO PARA EXCLUIR A VENDA COM BASE NO ID
def excluir_venda(id_venda):
  lista_remove = []
  for ids in TABELA_FATO:
    if ids['ID_VENDA'] == id_venda:
      lista_remove.append(ids)
  for i in lista_remove:
    TABELA_FATO.remove(i)
    #Além de excluir, adicionamos essa exclusão na lista de vendas excluidas
    VENDAS_EXCLUIDAS.append(i)
  return





#FUNCAO PARA CADASTRAR O VENDEDOR
def cadastra_vendedor():
  nome = input('Digite o nome do vendedor: \n\n')

  if nome in [lista_nomes[1] for lista_nomes in [itens for itens in DIM_VENDEDOR.items()]]:
    print('Vendedor Já Cadastrado')
    index = [id for id in [itens for itens in DIM_VENDEDOR.items()] if id[1] == nome][0][0]
  else:
    index = [lista_nomes[0] for lista_nomes in [itens for itens in DIM_VENDEDOR.items()]][-1] +1

  DIM_VENDEDOR[index]=nome

  return index

#FUNCAO PARA PEGAR O CPF, COM TRATAMENTO DE ERROS.
#Já EXISTE A FUNCAO PEGA_CLIENTE mas a ideia do PEGA_CLIENTE é a de checar existencia do cliente
#em conjunto com a função VALIDA_CLIENTE... o pega_cpf ta focado no cadastro. Trabalham em conjunto
#achei melhor separar para não colocar muita exceção no PEGA_CLIENTE.
def pega_cpf():

  try:
    cpf = int(input('Digite o CPF do Cliente \n\n'))
  except:
    print('Digitar o CPF em algarismos sem o . e o - ')
    cpf = pega_cpf()

  if len(str(cpf)) <= 11:
    cpf = str('{:0>11}'.format(cpf))

  else:
    print('O CPF deve conter apenas 11 digitos.')
    cpf = pega_cpf()


  return cpf

#FUNCAO PARA PEGAR O VENDEDOR, COM TRATAMENTO DE ERROS.
def cadastra_cliente():
  cpf = pega_cpf()
  if cpf not in DIM_CLIENTE:
     DIM_CLIENTE[cpf] = input('Digite o Nome do cliente: \n\n')
  else:
    print('Cliente já cadastrado')

  return cpf

#FUNCAO PARA PEGAR O VALOR UNITARIO DO PRODUTO, COM TRATAMENTO DE ERROS.
def pega_valor_UNIT():
  try:
    valor = float(input('Digite o preço unitário do produto'))
  except:
    print('Digite um preço Válido (separador decimal é o .)')
    valor = pega_valor_UNIT()
  if valor <=0:
    print('Favor digitar um valor maior do que 0')
    valor = pega_valor_UNIT()
  return valor

#FUNCAO PARA CADASTRAR O PRODUTO.
def cadastra_produto():
  nome = input('Digite o nome do Produto: \n\n')

  if nome in [lista_nomes[1][0] for lista_nomes in [itens for itens in DIM_PRODUTO.items()]]:
    print('Produto Já Cadastrado')
    index = [id for id in [itens for itens in DIM_PRODUTO.items()] if id[1] == nome][0][0]

  else:
    index = [lista_nomes[0] for lista_nomes in [itens for itens in DIM_PRODUTO.items()]][-1] +1
    valor = pega_valor_UNIT()
    DIM_PRODUTO[index]=[nome, valor]

  return index
```

```python
from functools import reduce


#FUNCAO PARA RETORAR O TOTAL DO mes/ano E O TICKET MÉDIO
def total_ticket(ano, mes):
  soma = lambda a,b: a+b #para usar reduce

  lista_valor = [ids['QTD_VENDIDA']*DIM_PRODUTO[ids['ID_PRODUTO']][1] for ids in TABELA_FATO if ids['DTA_VENDA'].split(sep = '/')[1] == mes and ids['DTA_VENDA'].split(sep = '/')[2] == ano]

  qtd_vendas = len(set([ids['ID_VENDA'] for ids in TABELA_FATO if ids['DTA_VENDA'].split(sep = '/')[1] == mes and ids['DTA_VENDA'].split(sep = '/')[2] == ano]))
  total = reduce(soma, lista_valor) #aplicando reduce
  ticket = total/qtd_vendas
  if total == 0:
    print('Não houve vendas no periodo requisitado')
  else:
    print(f'Total = {total}, ticket médio = {ticket}')

#FUNCAO PARA RETORAR O VENDEDOR QUE VENDEU MAIS EM R$ NO mes/ano
def melhor_vendedor(ano, mes):
  dict_vendedor = {}
  for i in DIM_VENDEDOR:
    lista_valor = [ids['QTD_VENDIDA']*DIM_PRODUTO[ids['ID_PRODUTO']][1] for ids in TABELA_FATO if ids['DTA_VENDA'].split(sep = '/')[1] == mes and ids['DTA_VENDA'].split(sep = '/')[2] == ano and ids['ID_VENDEDOR'] == i]
    total = sum(lista_valor)
    dict_vendedor[i]=total

  vendedor = list(dict(sorted(dict_vendedor.items(), key=lambda item: item[1],reverse=True)).keys())[0]
  valor = dict_vendedor[vendedor]
  vendedor_nome = DIM_VENDEDOR[vendedor]
  if total == 0:
    print('Não houve vendas no periodo requisitado')
  else:
    print(f'O vendedor com mais vendas em {mes}/{ano} foi: {vendedor_nome} ', f'Ele vendeu: R$ {valor}', sep = '\n')

#FUNCAO PARA RETORAR O PRODUTO QUE VENDEU MAIS EM QTD NO mes/ano
def melhor_produto(ano, mes):
  dict_produto = {}
  for i in DIM_PRODUTO:
    lista_valor = [ids['ID_PRODUTO'] for ids in TABELA_FATO if ids['DTA_VENDA'].split(sep = '/')[1] == mes and ids['DTA_VENDA'].split(sep = '/')[2] == ano and ids['ID_PRODUTO'] == i]
    total = len(lista_valor) #contagem
    dict_produto[i]=total

  produto = list(dict(sorted(dict_produto.items(), key=lambda item: item[1],reverse=True)).keys())[0]
  valor = dict_produto[produto]
  produto_nome = DIM_PRODUTO[produto][0]
  if total == 0:
    print('Não houve vendas no periodo requisitado')
  else:
    print(f'O produto mais vendido em {mes}/{ano} foi: {produto_nome} ', f'Ele vendeu: {valor} unidades', sep = '\n')

#FUNCAO PARA SALVAR O RELATÓRIO DE VENDAS NO mes/ano
def salvar_vendas(ano, mes):
  #Criando uma funçao geradora para pegar as vendas da FATO
  vendas_periodo = (ids for ids in TABELA_FATO if ids['DTA_VENDA'].split(sep = '/')[1] == mes and ids['DTA_VENDA'].split(sep = '/')[2] == ano)


  arquivo = open(f'./vendas_{mes}_{ano}.txt', "w")
  headers = ['ID Venda','Produto','Qtd Vendida','Vendedor','Cliente','Data','Obs','Valor Total']
  arquivo.write(','.join(headers))
  arquivo.write('\n')
  arquivo.close()

  for i in vendas_periodo:
    arquivo = open(f'./vendas_{mes}_{ano}.txt', "a")
    lista_valores = list(i.values())
    valor_venda = DIM_PRODUTO[lista_valores[1]][1]*lista_valores[2]

    #substituindo IDs pelos nomes
    lista_valores[1] = DIM_PRODUTO[lista_valores[1]][0]
    lista_valores[3] = DIM_VENDEDOR[lista_valores[3]]
    lista_valores[4] = DIM_CLIENTE[lista_valores[4]]

    #Convertendo o que não for str para str a fim de salvar no arquivo
    lista_convertida = [str(item) for item in lista_valores]

    #escrevendo no arquivo e no fim de tudo pulando uma linha
    arquivo.write(','.join(lista_convertida))
    arquivo.write(f',{valor_venda} \n')
    arquivo.close()

#funcao para pegar o cambio... porém na função de salvar em USD ja deixei um cambio definido
#para o caso de não passarem um cambio
def pega_cambio():
  try:
    cambio = float(input('US$ 1,00 equivale a quantos R$? : '))
  except:
    print('Digite o valor do cambio corretamente, com . como separador decimal')
    pega_cambio()
  if cambio <= 0:
    print('Não digite valores negativos')
    cambio = pega_cambio()

  return cambio

#FUNCAO PARA SALVAR O RELATÓRIO DE VENDAS NO mes/ano em USD
def salvar_vendas_USD(ano, mes,cambio=4.79):
  #Criando uma funçao geradora para pegar as vendas da FATO
  vendas_periodo = (ids for ids in TABELA_FATO if ids['DTA_VENDA'].split(sep = '/')[1] == mes and ids['DTA_VENDA'].split(sep = '/')[2] == ano)


  arquivo = open(f'./vendas_{mes}_{ano}_USD.txt', "w")
  headers = ['ID Venda','Produto','Qtd Vendida','Vendedor','Cliente','Data','Obs','Valor Total']
  arquivo.write(','.join(headers))
  arquivo.write('\n')
  arquivo.close()

  for i in vendas_periodo:
    arquivo = open(f'./vendas_{mes}_{ano}_USD.txt',  "a")
    lista_valores = list(i.values())
    valor_venda_USD = round(DIM_PRODUTO[lista_valores[1]][1]*lista_valores[2] / cambio,2)

    #substituindo IDs pelos nomes
    lista_valores[1] = DIM_PRODUTO[lista_valores[1]][0]
    lista_valores[3] = DIM_VENDEDOR[lista_valores[3]]
    lista_valores[4] = DIM_CLIENTE[lista_valores[4]]

    #Convertendo o que não for str para str a fim de salvar no arquivo
    lista_convertida = [str(item) for item in lista_valores]

    #escrevendo no arquivo e no fim de tudo pulando uma linha
    arquivo.write(','.join(lista_convertida))
    arquivo.write(f',{valor_venda_USD} \n')
    arquivo.close()

#FUNCAO PARA SALVAR AS VENDAS CANCELADAS EM UM TABELA
def vendas_canceladas(ano, mes):
  #Criando uma funçao geradora para pegar as vendas da FATO
  vendas_periodo = (ids for ids in VENDAS_EXCLUIDAS if ids['DTA_VENDA'].split(sep = '/')[1] == mes and ids['DTA_VENDA'].split(sep = '/')[2] == ano)



  print (['ID Venda','Produto','Qtd Vendida','Vendedor','Cliente','Data','Obs','Valor Total'])


  for i in vendas_periodo:

    lista_valores = list(i.values())
    valor_venda = DIM_PRODUTO[lista_valores[1]][1]*lista_valores[2]

    #substituindo IDs pelos nomes
    lista_valores[1] = DIM_PRODUTO[lista_valores[1]][0]
    lista_valores[3] = DIM_VENDEDOR[lista_valores[3]]
    lista_valores[4] = DIM_CLIENTE[lista_valores[4]]

    #Convertendo o que não for str para str a fim de salvar no arquivo
    lista_convertida = [str(item) for item in lista_valores]

    #escrevendo no arquivo e no fim de tudo pulando uma linha
    print(','.join(lista_convertida),f',{valor_venda} \n')

#FUNCAO DE MENU SECUNDÁRIO À MAIN PARA QUANDO O USUÁRIO SELECIONAR A OPCAO ANALISE VENDAS NA MAIN
def analise_venda():
  Valida_init = True
  while Valida_init == True:
    print('*'*71)
    print('*'*71, end = '\n\n')
    print('O que deseja fazer?\n\n')
    analise_on = input(' Valor total e Ticket médio por período - 1 \n Melhor Vendedor do Período - 2 \n Produto Mais Vendido do período - 3 \n Salvar vendas totais do período - 4 \n Salvar vendas totais do período em Dolar - 5 \n Consultar vendas Canceladas - 6 \n SAIR - 7 \n\n')
    print('*'*71)
    print('*'*71, end = '\n\n')
    #checando se foi digitado apenas as opções passadas
    if analise_on not in ('1','2','3','4','5','6','7'):
      print('Favor digitar uma das upções fornecidas \n\n')
    #ok
    elif analise_on == '7':
      Valida_init = False


    elif analise_on == '1':
      ano = str(pega_ano())
      mes = str(pega_mes())
      total_ticket(ano, mes)

    elif analise_on == '2':
      ano = str(pega_ano())
      mes = str(pega_mes())
      melhor_vendedor(ano, mes)

    elif analise_on == '3':

      ano = str(pega_ano())
      mes = str(pega_mes())
      melhor_produto(ano, mes)


    elif analise_on == '4':

      ano = str(pega_ano())
      mes = str(pega_mes())
      salvar_vendas(ano, mes)

    elif analise_on == '5':
      ano = str(pega_ano())
      mes = str(pega_mes())
      cambio = pega_cambio()
      salvar_vendas_USD(ano, mes, cambio)

    elif analise_on == '6':

      ano = str(pega_ano())
      mes = str(pega_mes())
      vendas_canceladas(ano, mes)

```
